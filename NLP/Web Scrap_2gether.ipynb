## Webscraping for 2gether

```python
# Importing needed libraries
import requests
from bs4 import BeautifulSoup
```

```python
# Pointing to URL
url= 'https://en.wikipedia.org/wiki/2gether:_The_Series_(Thai_TV_series)'
```

```python
# Sending request to get url
response= requests.get(url)
```

```python
# Checking the response
response # Positive response
```

```python
# Initializing parser
soup= BeautifulSoup(response.text, 'html.parser')
```

```python
# Check the Soap variable
soup
```

```python
# Fetching div "class -mw-parser-output" data
content = soup.find('div', {"class":"mw-parser-output"})
content
```

```python
# Creating the text data
synopsis= ''

for i in content.findAll("p"):
    synopsis= synopsis + i.text
    
print(synopsis)    
```
