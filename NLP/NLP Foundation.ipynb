## Text Preprocessing in NLP

```python
import re
import nltk
import numpy as np

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
```

```python
nltk.download('wordnet')
```

```python
stop_words= set(stopwords.words('english'))
len(stop_words)
```

```python
lemm = WordNetLemmatizer()
lemm
```

```python
texts = [
    "I love this product!!! 😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "Visit http://example.com now!",
    "This is AMAZING!!!"
]
```

```python
def clean_text(text):
    text= text.lower()
    text= re.sub(r'http\S+', '', text)
    text= re.sub(r'\d+', '', text)
    text= re.sub(r'[^\w\s]', '', text)
    text= re.sub(r'\s+', ' ', text)
    words= text.split()
    words= [w for w in words if w not in stop_words]
    words= [lemm.lemmatize(w) for w in words]
    return " ".join(words)
```

```python
cleaned= [clean_text(t) for t in texts]
cleaned
```

```python
for i,j in zip(texts, cleaned):
    print(f"Original: {i}")
    print(f"Cleaned: {j}")
```

## BoW and N-grams

Real Understanding (MOST IMPORTANT)

👉 BoW = Count of words in fixed vocabulary
👉 N-grams = add context
👉 Vector = numerical representation of text

⚠️ Limitations of BoW
No importance weighting ❌
Common words dominate ❌
No deep meaning ❌


What “No Importance Weighting” Means
BoW does NOT distinguish between:

    important words
    common/useless words
👉 It just counts everything blindly   

👉 So it assumes:
More frequency = more important

### Code (BoW)

```python
texts = ["I love NLP", "I love ML", "I hate NLP"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
```

```python
# "I" is missing because CountVectorizer by default ignores words with less than 2 characters.
# CountVectorizer sorts vocabulary alphabetically, not based on sequence.
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

### Add N-grams to BoW

```python
texts = ["I love love NLP", "I love ML", "I hate NLP"]
```

```python
vectorizer = CountVectorizer(ngram_range=(1,2)) # unigram + bigram
X = vectorizer.fit_transform(texts)
```

```python
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

N-gram → feature generator
BoW → counting method
TF-IDF → weighting method

N-grams are used along with vectorization techniques like Bag of Words or TF-IDF to capture word sequences and improve context representation.

👉 BoW = counts (integers)      0, 1, 2, 3...
👉 TF-IDF = importance (decimals)     0.0 to 1.0 (or sometimes slightly higher depending on scaling)

## Text Classification Model

#### 1️⃣ Sample Dataset

```python
texts = [
    "I love this product",
    "This is amazing",
    "Worst service ever",
    "I hate this",
    "Very bad experience",
    "I really liked it"
]

labels = [1, 1, 0, 0, 0, 1]  
# 1 = Positive, 0 = Negative
```

#### 2️⃣ Cleaning Function

```python
stop_words= set(stopwords.words('english'))
len(stop_words)
lemm = WordNetLemmatizer()
lemm
```

```python
cleaned= [clean_text(t) for t in texts]
cleaned
```

#### 3️⃣ Convert Text → Numbers (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer(ngram_range=(1,2))
X= vectorizer.fit_transform(cleaned)
```

```python
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

#### 4️⃣ Train Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, labels)
```

```python
labels
```

#### 5️⃣ Make Prediction

```python
test = ["this product is bad"]

test_clean = [clean_text(t) for t in test]
test_vec = vectorizer.transform(test_clean)
test_vec.toarray()
```

```python
prediction = model.predict(test_vec)
prob = model.predict_proba(test_vec)
prediction, prob
```

#### Check important words learned

```python
feature_names= vectorizer.get_feature_names_out()
feature_names
```

```python
coeffs= model.coef_[0]
coeffs
```

```python
for word, coef in zip(feature_names, coeffs):
    print(f"{word} : {coef}")
# 👉 Positive coef → positive sentiment
# 👉 Negative coef → negative sentiment
```

Interview Answer (GOLD)

I preprocess text, convert it into numerical features using TF-IDF with n-grams, and train a Logistic Regression model for classification.

Using only bigrams can lose important single-word information, so it’s best to use a combination of unigrams and bigrams.

Both BoW and TF-IDF use a fixed vocabulary learned during training, and unseen words in test data are ignored.

## Embeddings

Embeddings are vector representations of words, and Word2Vec is one model used to generate those embeddings.

### Word2Vec in the easiest way possible using gensim

#### 1️⃣ Install (only once)

```python
pip install gensim
```

#### 2️⃣ Sample Data (very important format)

👉 Word2Vec needs tokenized sentences (list of words)

```python
sentences = [
    ["i", "love", "nlp"],
    ["i", "love", "machine", "learning"],
    ["nlp", "is", "fun"],
    ["machine", "learning", "is", "powerful"],
    ["i", "enjoy", "learning"]
]
```

#### 3️⃣ Train Word2Vec Model

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences,
    vector_size=50,   # size of embedding
    window=3,         # context window
    min_count=1,      # include all words
    workers=4
)
```

```python
model
```

#### 4️⃣ Get Word Embedding

```python
model.wv["love"]
```

#### 5️⃣ Find Similar Words (MOST IMPORTANT 🔥)

```python
model.wv.most_similar("love")
```

```python
model.wv.most_similar("hi")
```

#### 6️⃣ Check similarity between words

```python
model.wv.similarity("love", "enjoy")
```

#### 7️⃣ Convert sentence → vector

```python
model.wv
```

```python
import numpy as np

def sentence_vector(sentence):
    vectors = [model.wv[word] for word in sentence if word in model.wv]
    return np.mean(vectors, axis=0)

vec = sentence_vector(["i", "love", "nlp"])
print(vec)
```

## Sentence embeddings and Semantic similarity 

### Sentence Transformers Code

```python
# pip install sentence-transformers
```

```python
sentences = [
    "I love machine learning",
    "I enjoy machine learning",
    "I hate machine learning"
]
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# It loads a pre-trained sentence embedding model
model= SentenceTransformer('all-MiniLM-L6-v2') 
model
```

```python
embeddings = model.encode(["I love NLP"])
len(embeddings)
```

```python
embeddings
```

```python
print(embeddings.shape)
# Yes, 384 represents the number of features (dimensions)/ columns in the embedding vector for each sentence.
# Dimension = model architecture choice
# 384 = vector size of model, Each sentence → 384 features
```

```python
embeddings = model.encode(sentences)
len(embeddings)
```

```python
embeddings
```

```python
sentences
```

```python
# similarity between sentence 1 and others
sim= cosine_similarity([embeddings[0]], embeddings)
sim
```

```python
# similarity between all
sim_matrix= cosine_similarity(embeddings)
sim_matrix
```

#### Test

```python
test = ["I hate this product"]
test_embedding= model.encode(test)
sim = cosine_similarity(test_embedding, embeddings)
sim
```

```python
best_ind= np.argmax(sim)
sentences[best_ind]
```

#### Find most similar sentence

```python
query = "I like machine learning"
query_emb = model.encode([query])
query_emb
```

```python
similarities = cosine_similarity(query_emb, embeddings)
similarities
```

```python
best_idx= np.argmax(similarities)
best_idx
```

```python
print("Most similar sentence:", sentences[best_idx])
```
