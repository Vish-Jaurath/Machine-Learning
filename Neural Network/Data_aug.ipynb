```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
img= load_img('jimin.jpg', target_size=(180,150))
img  
```

```python
x= img_to_array(img)
x
```

```python
x.shape
```

```python
x= x.reshape((1,) + x.shape)
x.shape
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
i=0
for j in datagen.flow(x, batch_size=1, save_to_dir='preview', save_prefix='Park', save_format='jpg'):
    
    i+=1
    if i> 19:
        break
```
